```python
import json
import pandas as pd
```

```python
with open('sample.json') as json_file:
    json_data = json.load(json_file)
```

```python
type(json_data)
```

```python
list(json_data.keys())
```

```python
for local_name in list(json_data.keys()):
    local_data = json_data[local_name]
    local_df = pd.DataFrame(local_data).T
    local_df.plot(figsize=(10,3))
```
